## Import the libraries

```python
import pandas as pd
import numpy as np
```

## Importing the dataset

```python
dataset = pd.read_csv('data.csv', header=0)
dataset.head()
```

### Cleaning data

```python
dataset.drop("id", axis=1, inplace=True)
dataset.drop("Unnamed: 32", axis=1, inplace=True) #remove unnamed column
```

Show unique values of diagnosis column:



```python
dataset.diagnosis.unique()
```

And map them to 1s and 0s:

```python
dataset['diagnosis'] = dataset['diagnosis'].map({'M':1, 'B':0})
```

```python
dataset.head()
```

## Data exploration

```python
dataset.describe()
```

```python
plt.hist(dataset['diagnosis'])
plt.title('Diagnosis: Malignant=1, Benign=0')
plt.show()
```

```python
malignant_data = dataset[dataset.diagnosis == 1]
benign_data = dataset[dataset.diagnosis == 0]
```

```python
malignant_data.head()
```

Create a list of features means:

```python
feature_mean = list(dataset.columns[1:11])
print(feature_mean)
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(8,10))
axes = axes.ravel()
for idx, ax in enumerate(axes):
  ax.figure
  binwidth = (max(dataset[feature_mean[idx]]) - min(dataset[feature_mean[idx]]))/50
  ax.hist([malignant_data[feature_mean[idx]], benign_data[feature_mean[idx]]],
          bins=np.arange(min(dataset[feature_mean[idx]]), max(dataset[feature_mean[idx]])+binwidth, binwidth),
          alpha=0.5,
          stacked=True,
          label=['Malignant', 'Benign'],
          color=['r', 'g'])
  ax.legend(loc='upper right')
  ax.set_title(feature_mean[idx])
plt.tight_layout()
plt.show()
```

## Model

Split dataset in train and test datasets:

```python
train_ds,  test_ds = train_test_split(dataset, test_size=0.3)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.metrics import confusion_matrix
```

```python
def model_classification(model, x_train, y_train, x_test, y_test):

  model.fit(x_train, y_train)

  y_pred = model.predict(x_test)
  cm = confusion_matrix(y_test, y_pred)

  print("Confusion Matrix:")
  print(cm)

  cm_accuracy = (cm[0][0] + cm[1][1]) / (cm[0][0] + cm[0][1] + cm[1][0] + cm[1][1])
  print("Accuracy: {:.3f}".format(cm_accuracy*100))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler

x_train = train_ds[feature_mean]
y_train = train_ds['diagnosis']

x_test = test_ds[feature_mean]
y_test = test_ds['diagnosis']

min_max_scaler = MinMaxScaler()
min_max_scaler.fit(x_train)
x_train = min_max_scaler.transform(x_train)
x_test = min_max_scaler.transform(x_test)

mdl = model_classification(LogisticRegression(), x_train, y_train, x_test, y_test)
```
